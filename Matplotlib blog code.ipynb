```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the dataset
df = pd.read_csv('merc 2.csv')
```

```python
#Take first 500 rows from df
df = df.head(500)
```

```python
df.head()
```

Matplotlib

```python
#Scatter plot
fig, ax = plt.subplots(figsize=(10,8))
#year against price
ax.scatter(df['year'], df['price'])
#add title and labels
ax.set_title('Mercedes Benz dataset')
ax.set_xlabel('Year')
ax.set_ylabel('Price')

plt.show()
```

```python
df.transmission.value_counts()
```

```python
#Scatter plot with different colours
colors = {'Semi-Auto':'r', 'Automatic':'g', 'Manual':'b'}

fig,ax = plt.subplots(figsize=(10,8))
for i in range(len(df['year'])):
    ax.scatter(df['year'][i], df['price'][i], color=colors[df['transmission'][i]])
ax.set_title('Mercedes Benz dataset')
ax.set_xlabel('Year')
ax.set_ylabel('Price')
plt.show()
```

```python
df.head()
```

```python
#Line chart
columns = df[['mpg','engineSize']]

x_data = range(0, df.shape[0])

fig,ax = plt.subplots(figsize=(10,8))
for column in columns:
    ax.plot(x_data, df[column], label=column)
    
ax.set_title('Mercedes Benz dataset')
ax.legend()
plt.show()
```

```python
#Histogram
fig,ax = plt.subplots(figsize=(10,8))
ax.hist(df['model'])
ax.set_title('Model distribution in this dataset')
ax.set_xlabel('Model')
ax.set_ylabel('Frequency')
plt.xticks(fontsize=10, rotation=25)
plt.show()
```

```python
#bar chart
fig,ax = plt.subplots(figsize=(10,8))
# count the occurence of each class
data = df['model'].value_counts().sort_index()

#x axis
model = data.index
#y axis
freq = data.values
ax.bar(model,freq)
ax.set_title('Model distribution')
ax.set_xlabel('Model')
ax.set_ylabel('Frequency')
plt.xticks(fontsize=10, rotation=25)
plt.show()
```

Seaborn

```python
import seaborn as sns
sns.set_theme(rc={'figure.figsize':(10,8)})
```

```python
#Scatter plot
sns.scatterplot(x='year', y='price', data=df).set_title('Mercedes Benz dataset')
```

```python
#Scatter plot with colours
sns.scatterplot(x='year', y='price', hue='transmission', data=df).set_title('Mercedes Benz dataset')
```

```python
#line chart
sns.lineplot(data=df[['mpg','engineSize']])
```

```python
#histogram
sns.histplot(df['model'], kde=False)
plt.xticks(fontsize=10, rotation=25)
plt.show()
```

```python
#histogram with kde
sns.histplot(df['model'], kde=True)
plt.xticks(fontsize=10, rotation=25)
plt.show()
```

```python
#Bar chart
sns.countplot(x=df['model'].sort_values()).set_title('Model distribution')
plt.xticks(fontsize=10, rotation=25)
plt.show()
```

```python
#Extra
#boxplot
sns.boxplot(x='transmission', y='price', data=df)
```

```python
sns.heatmap(df.corr(),annot=True)
```
